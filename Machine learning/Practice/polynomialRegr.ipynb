```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# Generate a simple dataset (example: quadratic relationship)
# x: Independent variable, y: Dependent variable (quadratic relationship)
x = np.random.uniform(1, 10, 100).reshape(-1, 1)  # Generate 100 random data points
y = 2 + 3 * x + 0.5 * x**2 + np.random.randn(100, 1) * 2  # Quadratic relationship with noise

# Visualize the original data
plt.scatter(x, y, color='blue', label='Original Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original Data')
plt.legend()
plt.show()

# Step 1: Create Polynomial Features (degree = 2 for quadratic)
poly = PolynomialFeatures(degree=2)  # For a quadratic fit
x_poly = poly.fit_transform(x)

# Step 2: Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.2, random_state=42)

# Step 3: Fit the Polynomial Regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Step 4: Make Predictions
y_pred = model.predict(x_test)

# Step 5: Visualize the Polynomial Fit
plt.scatter(x, y, color='blue', label='Original Data')  # Plot original data
x_range = np.linspace(min(x), max(x), 1000).reshape(-1, 1)  # Create a smoother line for the prediction
x_range_poly = poly.transform(x_range)  # Transform to polynomial features
y_range_pred = model.predict(x_range_poly)  # Predict using the trained model

plt.plot(x_range, y_range_pred, color='red', label='Polynomial Regression Fit')  # Plot polynomial fit
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()

# Step 6: Evaluate Model Performance
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```
